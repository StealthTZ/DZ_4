```python
import json
import random

# делаем файл на 100 значений заполненный случайными числами от 1 до 1000 в ормате json
data = [random.randint(1, 1000) for _ in range(100)]
with open('f_data.json', 'w') as file:
    json.dump(data, file)

print("готов файлик 'f_data.json'")
```

```python
# подгружаем данные
with open('f_data.json', 'r') as file:
    loaded_data = json.load(file)
# считаем среднее
    average = sum(loaded_data) / len(loaded_data)
    print("Среднее значение чисел в loaded_data:", average)
```

```python
# загружаем
with open('f_data.json', 'r') as file:
    loaded_data = json.load(file)

# преобразуем числа в текст
string_data = [str(num) for num in loaded_data]

print("Преобразованные данные в виде строк:")
print(string_data)
```

```python
# а теперь обратно в числа
to_numbers = [int(item) for item in string_data]

print("Обратное преобразование строк обратно в числа:")
print(to_numbers)
```

```python
import matplotlib.pyplot as plt

# Построение графика
plt.hist(loaded_data, bins=20, color='red')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.grid(True)
plt.show()

plt.plot(loaded_data, marker='o', color='green', linestyle='-')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.grid(True)
plt.show()

plt.bar(range(len(loaded_data)), loaded_data, color='orange')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.grid(axis='y')
plt.show()
```
