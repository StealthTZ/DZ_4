```python
conn = psycopg2.connect(
    dbname="postgres",
    user="db_user",
    password="12345678",
    host="localhost",
    port="5432")

cursor = conn.cursor()

# вычисляем сумму значений столбца
cursor.execute('SELECT SUM(value) FROM numbers')
sum_result = cursor.fetchone()[0]
print("Сумма значений столбца 'value':", sum_result)

# вычисляем среднее значения столбца
cursor.execute('SELECT AVG(value) FROM numbers')
avg_result = cursor.fetchone()[0]
print("Среднее значение столбца 'value':", avg_result)

cursor.close()
conn.close()
```

```python
conn = psycopg2.connect(
    dbname="postgres",
    user="db_user",
    password="12345678",
    host="localhost",
    port="5432")

cursor = conn.cursor()

# создаем вью содержащее только четные числа
create_view_query = '''
CREATE VIEW even_numbers_view AS
SELECT * FROM numbers WHERE value % 2 = 0
'''
cursor.execute(create_view_query)
conn.commit()

# получаем данные в переменную
cursor.execute('SELECT * FROM even_numbers_view')
rows = cursor.fetchall()

# печатаем 
for row in rows:
    print(row)

cursor.close()
conn.close()
```

```python
conn = psycopg2.connect(
    dbname="postgres",
    user="db_user",
    password="12345678",
    host="localhost",
    port="5432")

cursor = conn.cursor()

# собираем две таблицы в вью
query = '''
    CREATE OR REPLACE VIEW joined_view AS
    SELECT n1.id AS id_1, n1.value AS value_1, n2.id AS id_2, n2.value AS value_2
    FROM numbers AS n1
    JOIN numbers_01 AS n2 ON n1.id = n2.id;
'''

cursor.execute(query)
conn.commit()

# Выполнение запроса для вывода содержимого представления
cursor.execute('SELECT * FROM joined_view')

# Получение результатов запроса
rows = cursor.fetchall()

# Вывод результатов на экран
for row in rows:
    print(row)

cursor.close()
conn.close()
```
